```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 = pd.read_csv("constituency_1.csv")
df_2 = pd.read_csv("constituency_2.csv")
df_3 = pd.read_csv("constituency_3.csv")
df_4 = pd.read_csv("constituency_4.csv")
df_5 = pd.read_csv("constituency_5.csv")
df_6 = pd.read_csv("constituency_6.csv")
df_7 = pd.read_csv("constituency_7.csv")
df_8 = pd.read_csv("constituency_8.csv")
df_9 = pd.read_csv("constituency_9.csv")
df_10 = pd.read_csv("constituency_10.csv")
```

```python
old_party = ["Nepali Congress", "CPN UML" , "CPN Maxist Leninist", "Nepali Congress (BP)"]
```

```python
df_1.head(20)
```

```python
df = [df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10]
no_of_candidates = []
no_of_individual_parties = []
no_of_independent_candidates = []
for i, df_i in enumerate(df, start=1): 
    no_of_candidates.append(df_i["party"].value_counts().sum())
    no_of_individual_parties.append(df_i["party"].nunique())
    no_of_independent_candidates.append((df_i["party"] == "Independent").sum())
```

```python
# Creating labels for constituencies
constituencies = [f"C_{i+1}" for i in range(len(no_of_independent_candidates))]
```

# Visualization

```python
# NO of Candidates in each Constituency
plt.figure(figsize=(10,10))
sns.barplot(x=constituencies,y=no_of_candidates)
plt.title("Number of Candidates per Constituency")
plt.xlabel("Constituency")
plt.ylabel("Number of Candidates")
plt.xticks(rotation=45)
plt.show()
```

```python
# NO of Independent candidates compitating in each Constituency
plt.figure(figsize=(10,10))
sns.barplot(x=constituencies,y=no_of_independent_candidates)
plt.title("Number of Independent candidates competing per Constituency")
plt.xlabel("Constituency")
plt.ylabel("Number of Individual Candidates")
plt.xticks(rotation=45)
plt.show()
```

```python
Percentage_of_individual_candidates = []
for ind, total in zip(no_of_independent_candidates,no_of_candidates):
    Percentage_of_individual_candidates.append((ind/total)*100)
```

```python
print(f"On average {sum(Percentage_of_individual_candidates)/len(Percentage_of_individual_candidates)} of candidates are the individual candidates")
```
